**DATA CLEANING PROJECT - MARKETING CAMPAIGN**

```python
# Import Required Libraries
import pandas as pd
import numpy as np
```

**Loading Dataset**

```python
ile_path = "/mnt/data/marketing_campaign.csv"
df = pd.read_csv("/content/marketing_campaign.csv", )
print(" Dataset Loaded Successfully!")
print("Shape of dataset:", df.shape)
```

**Basic Data ExplorationBasic Data Exploration**

```python
print("\nFirst 5 Rows:")
print(df.head())
print("\n Dataset Info:")
print(df.info())
print("\n Missing Values:")
print(df.isnull().sum())
```

```python
#Handle Missing Values
# Fill numeric columns with median
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
# Fill categorical columns with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
print("\n Missing Values Handled")
```

**Remove Duplicate Rows**

```python
before_duplicates = df.shape[0]
df = df.drop_duplicates()
after_duplicates = df.shape[0]
print(f"\n Removed {before_duplicates - after_duplicates} Duplicate Rows")
```

```python
#  Standardize Text Formatting
for col in categorical_cols:
    df[col] = df[col].str.strip()      # Remove extra spaces
    df[col] = df[col].str.lower()      # Convert to lowercase
print("\n Text Columns Standardized")
```

```python
# Convert Date Columns (if present)
# Example: If dataset contains a column named 'Dt_Customer'
if 'Dt_Customer' in df.columns:
    df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], errors='coerce')
    print("\n Date Column Converted to Proper Datetime Format")
# Fix Data Types
# Example: Ensure 'Year_Birth' is integer
if 'Year_Birth' in df.columns:
    df['Year_Birth'] = df['Year_Birth'].astype(int)
print("\n Data Types Checked and Fixed")
#  Rename Columns (Clean Headers)
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print("\n Column Names Cleaned")
# Final Dataset Check
print("\n Final Dataset Info:")
print(df.info())
print("\n Final Missing Values:")
print(df.isnull().sum())
print("\n Final Shape:", df.shape)
#  Save Cleaned Dataset
output_file = "/content/marketing_campaign_cleaned.csv"
df.to_csv(output_file, index=False)
print(f"\n Cleaned dataset saved as: {output_file}")
```
